```python
# Libraries used
import pandas as pd
import numpy as np
import random
import decimal
import math
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# Declared Variables for Earth sized planet
EARTHMOO_time =[]
EARTHMOO_eccentricity = []
EARTHMOO_inclination = []
# Declared Variables for Jupiter sized planet
JUPITER_time = []
JUPITER_eccentricity = []
JUPITER_inclination = []
```

```python
# This turns the string into 
def Funky(stuff):
    count = 0
    for x in stuff:
        x = float(x)
        stuff[count] = x
        count = count + 1
    return stuff
```

```python
# Doing this for each run
def EARTHMOOdownloader(pathname):
    global EARTHMOO_time
    global EARTHMOO_eccentricity
    global EARTHMOO_inclination
    with open(pathname, 'r') as h:
        lines_after_header = h.readlines()[4:]
        for x in lines_after_header:
            EARTHMOO_time.append(x.split()[0])
            EARTHMOO_eccentricity.append(x.split()[2])
            EARTHMOO_inclination.append(x.split()[3])
        h.close()
    EARTHMOO_time = Funky(EARTHMOO_time)
    EARTHMOO_eccentricity = Funky(EARTHMOO_eccentricity)
    EARTHMOO_inclination = Funky(EARTHMOO_inclination)
    
```

```python
# Doing this for each run
def JUPITERdownloader(pathname):
    global JUPITER_time
    global JUPITER_eccentricity
    global JUPITER_inclination
    with open(pathname, 'r') as h:
        lines_after_header = h.readlines()[4:]
        for x in lines_after_header:
            JUPITER_time.append(x.split()[0])
            JUPITER_eccentricity.append(x.split()[2])
            JUPITER_inclination.append(x.split()[3])
        h.close()
    JUPITER_time = Funky(EARTHMOO_time)
    JUPITER_eccentricity = Funky(EARTHMOO_eccentricity)
    JUPITER_inclination = Funky(EARTHMOO_inclination)
    
```

```python
# Reset each of the arrays
def Resetter():
    global EARTHMOO_time
    global EARTHMOO_eccentricity
    global EARTHMOO_inclination
    global JUPITER_time
    global JUPITER_eccentricity
    global JUPITER_inclination
    EARTHMOO_time = []
    EARTHMOO_eccentricity = []
    EARTHMOO_inclination = []
    JUPITER_time = []
    JUPITER_eccentricity = []
    JUPITER_inclination = []
```

```python
# Graphing the eccentricity vs. time
def GraphingEccentricity(PlotSaveName):
    plt.plot(EARTHMOO_time, EARTHMOO_eccentricity,'g')
    plt.plot(JUPITER_time, JUPITER_eccentricity, 'r')
    plt.title('Eccentricity v. Time')
    plt.ylabel('e')
    plt.xlabel('Time')
    plt.savefig(PlotSaveName, format='eps', dpi=1000)
```
