```python
import numpy as np
import random
import decimal
import math

def Funky(stuff):
    count = 0
    for x in stuff:
        x = float(x)
        stuff[count] = x
        count = count + 1
    return stuff
    
    
```

```python
time = [] #Very first Column
a = [] #This is the second column
e = [] #This is the third column
i = [] #This is the fourth column
time2 = []
a2 = []
e2 = []
i2 = []
```

```python
with open('/Users/becerrilblas/Documents/Martin Group Research/4SIMS/run3/JUPITER.aei', 'r') as h:
    lines_after_Juan = h.readlines()[4:]
    for x in lines_after_Juan:
        time.append(x.split()[0])
        a.append(x.split()[1])
        e.append(x.split()[2])
        i.append(x.split()[3])
    h.close()
    
time = Funky(time)
a = Funky(a)
e = Funky(e)
i = Funky(i)
```
